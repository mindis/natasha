# Пример извлечения имён из отзывов об учителях школ и воспитателях детских садов

В Москве около 700 школ, у каждой есть сайт на портале mskobr.ru. На каждом сайте есть раздел с отзывами, например, http://lycc1501.mskobr.ru/elektronnye_servisy/reviews/ . Чаще всего там публикуют благодарности воспитателям детских садиков. Наташей можно извлекать упоминания сотрудников и сортировать их по числу благодарностей, как это сделано на http://obr.msk.ru/sch1329-sad.html:
<img src="https://habrastorage.org/files/726/6b9/eb3/7266b9eb3bd349e18c14afabc69adaf1" />

```python
lines = []
with open('reviews.txt') as file:
    for line in file:
        line = line.rstrip('\n')
        lines.append(line)
```

Вручную проверим качество разметки. Всего 378 метчей, 5 ошибок и 17 полуошибок (части имени расклеились, часть имени не нашлась) — качество ~6%.

Некоторые ошибки:
1. Сложные фамилии: "Мацаль Елене", "Лагуненок Валентине", "Войтик Марине", "А. Дистервег"
2. Не очень сложный фамилии: "Прусаковой Татьяны Николаевны", "Цаплина Елена", "Евгений Филькин"
3. "Деда Мороза"

```python
from random import sample
from random import seed as random_seed

from natasha import NamesExtractor

from IPython.display import display

extractor = NamesExtractor()
random_seed(41)
for text in sample(lines, 100):
    matches = extractor(text)
    display(matches)
    for index, match in enumerate(matches):
        print(index, match.fact)
```
