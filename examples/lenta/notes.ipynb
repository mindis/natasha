# Пример извлечения упоминаний персон их новостей

```python
texts = []
with open('news.txt') as file:
    for line in file:
        texts.append(line.rstrip())
```

```python
from random import sample
from random import seed as random_seed

from IPython.display import display

from natasha import PersonExtractor


extractor = PersonExtractor() 

random_seed(43)
count = 0
for text in sample(texts, 100):
    matches = extractor(text)
    display(matches)
```
