# Пример извлечения адресов из комментариев к ПЗЗ

Параллельно с историей про реконструкцию в Москве обсуждались новые Правила землепользования и застройки (ПЗЗ). Был проведён опрос населения. Более 100 000 комментариев были выложены в открытый доступ в виде огромного pdf-файла http://telegra.ph/PZZ---analitika-proshedshih-slushanij-05-01. Наташей можно извлечь упомянутые адреса и посмотреть в каких районах закон поддерживают, а в каких нет:
<img src="https://habrastorage.org/web/856/518/44a/85651844a0074f1aa5a25cc14d82c29f.png"/>

```python
texts = []
with open('comments.txt') as file:
    for line in file:
        line = line.rstrip()
        texts.append(line)
```

```python
from IPython.display import display

from tqdm import tqdm_notebook as log_progress

from natasha import AddressExtractor

extractor = AddressExtractor() 

for text in log_progress(texts):
    matches = extractor(text)
    if matches:
        display(matches)
```
